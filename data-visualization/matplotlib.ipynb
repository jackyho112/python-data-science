```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
new_axes = fig.add_axes([0,0,1,1])
new_axes.plot(x, y, 'r')
new_axes.set_xlabel('x')
new_axes.set_ylabel('y')
new_axes.set_title('title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x, y, 'r')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_ylim([0, 10000])
ax1.set_xlim([0, 100])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_ylim([30, 50])
ax2.set_xlim([20, 22])
ax1.plot(x, z, 'r')
ax2.plot(x, y, 'r')
ax2.set_title('zoom')
```

```python
plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, color='blue', lw=2, ls='--')
axes[1].plot(x, z, color='purple', lw=5, ls='-')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,3))
axes[0].plot(x, y, color='blue', lw=2, ls='--')
axes[1].plot(x, z, color='purple', lw=5, ls='-')
```
