```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic,kind='scatter')
```

```python
sns.distplot(titanic['fare'], kde=False, bins=30, color='red')
```

```python
sns.boxplot(x="class", y="age", data=titanic, palette='rainbow')
```

```python
sns.swarmplot(x="class", y="age", data=titanic, palette='Set2')
```

```python
sns.countplot(x="sex", data=titanic)
```

```python
ax = sns.heatmap(titanic.corr(), annot=True, cmap='coolwarm')
ax.set_title('titanic.corr()')
```

```python
g = sns.FacetGrid(titanic, col="sex")
g = g.map(plt.hist, "age", ec='black')
```
